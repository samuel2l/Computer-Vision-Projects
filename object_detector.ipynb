```python
import cv2

# Set the threshold to detect objects
thres = 0.45  

# Initialize video capture
cap = cv2.VideoCapture(0)
cap.set(3, 1280)  # Set width
cap.set(4, 720)   # Set height
cap.set(10, 70)   # Set brightness (optional)

# Load class names from coco.names file
classNames = []
classFile = '/Users/samuel/realtime_obj_detection/coco.names'
with open(classFile, 'rt') as f:
    classNames = f.read().rstrip('\n').split('\n')  # Correct split by '\n'

print(classNames)  # Print loaded class names

# Load the configuration and weights for the model
configPath = '/Users/samuel/realtime_obj_detection/ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
weightsPath = '/Users/samuel/realtime_obj_detection/frozen_inference_graph.pb'

# Create a DNN detection model
net = cv2.dnn_DetectionModel(weightsPath, configPath)
net.setInputSize(320, 320)
net.setInputScale(1.0 / 127.5)
net.setInputMean((127.5, 127.5, 127.5))
net.setInputSwapRB(True)

while True:
    success, img = cap.read()  # Read the frame
    classIds, confs, bbox = net.detect(img, confThreshold=thres)

    if classIds is not None and len(classIds) > 0:
        for (classId, confidence), box in zip(zip(classIds.flatten(), confs.flatten()), bbox):
            if classId - 1 < len(classNames):
                cv2.rectangle(img, box, color=(0, 255, 0), thickness=2)
                cv2.putText(img, classNames[classId - 1].upper(), (box[0] + 10, box[1] + 30),
                            cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
                cv2.putText(img, str(round(confidence * 100, 2)), (box[0] + 200, box[1] + 30),
                            cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('o/p', img)  
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```
